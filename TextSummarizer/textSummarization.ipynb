```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
api_key = os.getenv('OPENAI_API_KEY')
llm = ChatOpenAI()
```

```python
llm
```

```python
from langchain.schema import ( AIMessage,HumanMessage,SystemMessage)
speech =""" 
India a great nation. It is the most populous country in the world. It is the argest democracy in the world. 
Its has over 1000 language dialects spoken. There are 29 states and 6 union territories.
It is the land of Yoga and Ayurveda.
One of the best singers in the contry come from the state of Karnataka.
Hindustani and Carnatic classical music is prominent .
"""
speech
```

```python
chat_message =[
    SystemMessage(content='You are an expert in summarizing speech'),
    HumanMessage(content=f'Summarize the speech {speech}')
]

llm.get_num_tokens(speech)
```

```python
llm(chat_message).content
```

```python
llm(chat_message)
```

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate


generic_template = """
Write a summary for the following speech {speech} and translate the summary to the language {language}
"""

prompt = PromptTemplate(
    input_variable=['speech','language'],
    template=generic_template
)

prompt
```

```python
complete_template = prompt.format(speech=speech,language='Kannada')
complete_template
```

```python
llm_chain = LLMChain(llm=llm,prompt=prompt)
summary = llm_chain.run({'speech':speech,'language':"Kannada"})
summary
```
